2. **Crowd avoidance**: given a destination and a month, recommend which are the 5 best days and time range (morning, afternoon, evening and night) where there is a low number of flights departing from the airport, so there will be less people on there. The time ranges will be: 
    * *Morning*:     5 am to 12 pm (noon) 
    * *Afternoon*:   12 pm to 5 pm
    * *Evening*:     5 pm to 9 pm
    * *Night*:       9 pm to 4 am



```python
# import pandas as pd
import pandas as pd
import numpy as np

# Takes the file's folder
filepath = r"C:\Users\danie\Desktop\WBS.DataScience_Bootcamp\Week_2_Python\Bouns_Challenge\LAflights.csv"

# read the CSV file
LAflights = pd.read_csv(filepath)

# prompt for dest and month
#dest_input = input("Please enter your 3 character destination code")
#month_input = input("Please enter you preferred month of travelling.")

# defining the function 'crowd_avoidance'
def crowd_avoidance(dest_input, month_input):
# datetime stringslicing in order to get daytime (creating a new column in DF)
    a = LAflights["fl_datetime"].str.split()
    flt = list() # empty list flt = flight_time
    fld = list() # empty lis fld = flight_date
    for x in a:
        fld.append(x[0]) # list of flight dates
        flt.append(x[1]) #
    LAflights["daytime"] = flt
    LAflights["fl_date"] = fld

# continued string-slicing to extract the month-value from the flight-date-list in order to handle the user input on a monthly basis, again setting up a new column
    flm = list()
    for y in fld:
        z = y.split("-")
        flm.append(z[1])
    LAflights["month"] = flm
    
# setting up time-ranges using the newly created daytime-column
    LAflights["time_range"] = (
        np.where(LAflights["daytime"] <= "05:00:00", "Night",
        np.where(LAflights["daytime"] <= "12:00:00", "Morning",
        np.where(LAflights["daytime"] <= "17:00:00", "Afternoon",
        np.where(LAflights["daytime"] <= "21:00:00", "Evening", "Night"))))
    )   

# filtering the DF according to the user-input for dest and month, saving the filtered DF as a new variable
    LAselected = LAflights[(LAflights["dest"] == dest_input) & (LAflights["month"] == month_input)]

# group the newly set-up DF by flight-date, time_range and origin, aggregate on count()tailnumber(?), again stored in a new variable
    LArecommendations = LAselected.groupby(by = ["fl_date", "origin", "time_range"]).agg({"tail_num": "count"}).sort_values("tail_num")
    return LArecommendations.head(5)
```
